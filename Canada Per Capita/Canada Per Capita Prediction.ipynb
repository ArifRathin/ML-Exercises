```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
train_data = pd.read_csv(r"F:\Machine Learning\Exercises\Canada Per Capita\canada_per_capita_income.csv")
```

```python
Y_train = train_data["per capita income (US$)"]
X_train = train_data.drop("per capita income (US$)",axis=1)
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(X_train,Y_train,color="red",marker="*")
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train,Y_train)
prediction = linear_regression.predict([[2020],[2021],[2022],[2023],[2024]])
print(prediction)
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter([[2020],[2021],[2022],[2023],[2024]],prediction,color="blue",marker="H")
plt.plot([[2020],[2021],[2022],[2023],[2024]],prediction,color="blue")
```

```python
df = pd.DataFrame({"year":[2020,2021,2022,2023,2024],"per capita income (US$)":prediction})
df.to_csv(r"F:\Machine Learning\Exercises\Canada Per Capita\predictions.csv",index=False)
```
